```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D, art3d
import numpy as np
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d.proj3d import proj_transform
import os.path as osp
from matplotlib.backends.backend_pdf import PdfPages

def do_3d_projection(self, renderer=None):
    if renderer is None:
        renderer = self.axes.figure.canvas.get_renderer()
    xs, ys, zs = self._verts3d
    xs, ys, zs = proj_transform(xs, ys, zs, self.axes.get_proj())
    self.set_positions((xs[0], ys[0]), (xs[1], ys[1]))
    return np.min(zs)

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D, art3d
import numpy as np
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d.proj3d import proj_transform

# Custom 3D arrow class with a default renderer in do_3d_projection.
class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        super().__init__((0, 0), (0, 0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        # Project the 3D coordinates to 2D
        xs, ys, zs = self._verts3d
        xs, ys, zs = proj_transform(xs, ys, zs, self.axes.get_proj())
        self.set_positions((xs[0], ys[0]), (xs[1], ys[1]))
        super().draw(renderer)

    def do_3d_projection(self, renderer=None):
        # if renderer is None:
        #     renderer = self.axes.figure.canvas.get_renderer()
        xs, ys, zs = self._verts3d
        xs, ys, zs = proj_transform(xs, ys, zs, self.axes.get_proj())
        self.set_positions((xs[0], ys[0]), (xs[1], ys[1]))
        return np.min(zs)

# Create a figure and a 3D axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the start and end points of the vector
x = [1, 2]
y = [0.5, 2]
z = [0.1, 3]

# Create an Arrow3D instance with a custom arrow head style.
arrow = Arrow3D(x, y, z, mutation_scale = 18, lw=4, arrowstyle="->", color="#FF8040" , shrinkA=0, shrinkB=0)
ax.add_artist(arrow)

# Set the limits of the axes for a better view
ax.set_xlim([0, 3.5])
ax.set_ylim([0, 3.5])
ax.set_zlim([0, 3.5])
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
# Optionally remove axis tick numbers and frames

ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
# ax.set_axis_off()
plt.tight_layout()

plt.savefig(osp.join('arrows', '1.svg'),format= "svg")
plt.show()
```
